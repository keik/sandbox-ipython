```python
import wave
import numpy as np
import matplotlib.pyplot as plt
import  IPython.display as disp
%matplotlib inline
```

## Read from WAV file

```python
f = wave.open('data/ambient-roald-velden-morning-glow-60-90s.wav', 'r')

frames_as_string = f.readframes(f.getnframes())
frames = np.fromstring(frames_as_string, dtype=np.int32)
rate = f.getframerate()
channels = f.getnchannels()
sample_width = f.getsampwidth()

print('frames: {}, frame rate: {}, channels: {}, sample width: {}'.format(len(frames), rate, channels, sample_width))
disp.display(disp.Audio(frames, rate=rate))
```

## Time series

```python
fig = plt.figure()
axes1 = fig.add_subplot(111)
axes1.plot(frames[:4096])
axes1.set_xlim(0, 4096)

plt.show()
```
